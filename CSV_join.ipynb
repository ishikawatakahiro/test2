```python
import pandas as pd
import glob
import os
import datetime

```

```python
# パスで指定したファイルの一覧をリスト形式で取得. （ここでは一階層下のtestファイル以下）

path=input("インポートするエクセルファイルのパスを指定して下さい。(csv,xlsx等可能、同じ拡張子のみ)")
path=path.replace("\\","/")
os.chdir(path+"/")
csv_files=glob.glob("*")
ex_path=input("エクスポート先のパスを指定して下さい")
ex_path=ex_path.replace("\\","/")
#読み込むファイルのリストを表示
# for a in csv_files:
#     print(a)

#csvファイルの中身を追加していくリストを用意
data_list = []

#読み込むファイルのリストを走査
for file in csv_files:
#     data_list.append(pd.read_csv(file))
    data_list.append(pd.read_excel(file))

#リストを全て行方向に結合
#axis=0:行方向に結合, sort
df = pd.concat(data_list, axis=0)
t_delta = datetime.timedelta(hours=9)
JST = datetime.timezone(t_delta, 'JST')
now_jst = datetime.datetime.now(JST)
now = now_jst.strftime('%Y%m%d%H%M%S')

df.to_csv(ex_path+"/Export_data_"+now+".csv",index=False,encoding="cp932")
```
